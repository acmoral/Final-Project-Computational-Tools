```
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
def gamma(v):
  return 1/math.sqrt(1-v**2)
def xprima(x,gamma,v,t):
  return gamma*(x-v*t)
def tprima(x,gamma,v,t):
  return gamma*(t-v*x) 
import random

number_of_colors = 24

```

```
V=3/5
g=gamma(V)
thetaprime=np.linspace(0,345*math.pi/180,23)
vx=(np.cos(thetaprime)+V)/(np.cos(thetaprime)*V+1)
vy=np.sin(thetaprime)*g*(1-V*vx)
vxp=np.cos(thetaprime)
vyp=np.sin(thetaprime)
theta=np.arctan(vy/vx)
whole=np.array([vx,vy,theta])
whole=np.transpose(whole)
df = pd.DataFrame(data=whole, index=thetaprime, columns=['vx','vy','theta'])
from IPython.display import display
display(df) 
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(14,7))
for i in range(23):
  col = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(number_of_colors)]
  ax1.plot([0,vx[i]],[0,vy[i]],color=col[i])
  ax2.plot([0,vxp[i]],[0,vyp[i]],color=col[i])
  ax1.text(vx[i],vy[i]+0.01,str('%s' % float('%.2g' % (theta[i]*180/math.pi))))
  ax2.text(vxp[i],vyp[i]+0.01,str('%s' % float('%.2g' % (thetaprime[i]*180/math.pi))))
ax1.set_title('Marco K',fontsize=15)
ax2.set_title('Marco $K^{\prime}$',fontsize=15)
ax1.set_xlabel('x',fontsize=15)
ax1.set_ylabel('y',fontsize=15)
```

#Viaje de alice 

```
#Primera parte del viaje,aceleracion propia 1año-1
tau=np.linspace(0,5,100)
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16, 8))
a0=1
t=1/a0*np.sinh(a0*tau)
x=(1/a0*np.cosh(a0*tau))-1/a0
xi=(1/a0*math.cosh(a0*5))-1/a0
ti=1/a0*math.sinh(a0*5)
ax1.plot(x,tau,color='orange')
#segunda parte del viaje, aceleracion propia negativa
tau1=np.linspace(10,5,100)
t1=-1/a0*np.sinh(10*a0-a0*tau1)+2*ti
x1=-1/a0*np.cosh(-a0*tau1+10*a0)+1/a0+2*xi
ax1.plot(x1,tau1,color='black')
#tercera parte
tau2=np.linspace(10,15,100)
t2=-1/a0*np.sinh(-(a0*tau2)+(10*a0))+2*ti
x2=-1/a0*np.cosh(-(a0*tau2)+(10*a0))+1/a0+2*xi
ax1.plot(x2,tau2,color='red')
#Cuarta parte
tau3=np.linspace(20,15,100)
t3=1/a0*np.sinh((a0*tau3)-(20*a0))+4*ti
x3=1/a0*np.cosh((a0*tau3)-(20*a0))-1/a0
ax1.plot(x3,tau3,color='blue')
ax1.set_xlabel('x',fontsize=15)
ax1.set_ylabel(r'$\tau$',fontsize=15)
#Pero ahora deberia servir para t no?
ax2.plot(x,t,color='orange')
ax2.plot(x1,t1,color='black')
ax2.plot(x2,t2,color='red')
ax2.plot(x3,t3,color='blue')
ax2.set_xlabel('x',fontsize=15)
ax2.set_ylabel('t',fontsize=15)
ax1.set_title(r'$\tau$ VS $1/a_0\cosh(a_0\tau)$')
ax2.set_title(r'$t$ VS $1/a_0\cosh(a_0\tau)$ Linea de Mundo de Alice en K')
```
